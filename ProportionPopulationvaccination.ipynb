```python
import pandas as pd
```

```python
data = pd.read_csv('cleanVaccinations.csv')
```

```python
# display content of data frame
data.head(5)
```

```python
# change date column to datetime type to get access to datetime functions
data['date'] = pd.to_datetime(data.date, dayfirst= True)
data['date'].head(5)
```

```python
data
```

```python
# date range 05-03-2021 to 15-03-2021
startdate, enddate = [pd.to_datetime('05/03/2021',dayfirst=True),pd.to_datetime('15/03/2021',dayfirst=True)]

# return dataframe in the range of start and end date
data = data.loc[(data['date'] >= startdate)&(data['date'] <= enddate)]
```

```python
# sums up columns
ndata = data.groupby(['country','Population']).sum()

ndata.head(6)
```

```python
# list on indexes containing countries and population
lst = [x for x in ndata.index.values]

# extract population and country from dataframe's index
countries = []
population = []

for x in range(len(lst)):
    countries.append(lst[x][0]) # country
    population.append(int(float(lst[x][1]))) # population
    
```

```python
# create new dataframe with the required data
pfv = ndata.reset_index(drop=True)['people_fully_vaccinated']
```

```python
# create new dataframe with only required data (country population and people fully vaccinated)
newdf = pd.DataFrame({'country':countries,'population':population,'people_fully_vaccinated':pfv})
```

```python
# calculate the rate of people fully vaccinated and create new column for rate
# R = (People fully vaccinated / population) / ×100

# create new column and assign the rate to it
newdf['rate'] = (newdf['people_fully_vaccinated'] / newdf['population'])*100
```

```python
# sort the dataframe by Rate in descending order
newdf.sort_values(by='rate', inplace=True, ignore_index=True, ascending=False)

newdf
```

#### As shown above, the data set cointains error were people fully vaccinated  are higher than the population for some countries (ex: Gibraltar and Israel)
#### and Zero (0) for others (ex: Zimbabwe and Azerbaijan)

```python
# filtered only required countries
countries = 'United Kingdom', 'Romania', 'Brazil', 'Bulgaria', 'United Arab Emirates'
required_countries_df = newdf.loc[newdf['country'].isin(countries)]

# preview required countries' ranking
required_countries_df
```

## Remove countries with rate over 100 and Zero(0) because the dataset currupted

```python
# remove Rate higher the 100 or 0 from data fram and present the remaining,
newdf = newdf.loc[(newdf.rate <= float(100)) & (newdf.rate > float(0))]
newdf
```

```python
# new ranking for our countries after removing Zeros and over 100 rates
# filtered only required countries
countries = 'United Kingdom', 'Romania', 'Brazil', 'Bulgaria', 'United Arab Emirates'
required_countries_df = newdf.loc[newdf['country'].isin(countries)]

# preview new required countries' ranking
# NOTE: the index starts from Zero(0), 13 means 14th place
required_countries_df
```

### note: UAE was romoved because its rate was 0

## Highest and Lowest Ranking

```python
### Lowest and highest ranking countries
print(f'Highest Ranking {newdf["country"].values[0]} - {newdf.rate.max()}%')  #highest ranking

print(f'Highest Ranking {newdf["country"].values[-1]} - {newdf.rate.min()}%') #lowest ranking
```
